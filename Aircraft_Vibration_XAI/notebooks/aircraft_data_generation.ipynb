```python
import numpy as np
import pandas as pd
np.random.seed(42)
n = 4000
data = []
for _ in range(1600):
    rpm = np.random.normal(20000,500)
    temp = np.random.normal(600,30)
    vib = np.random.normal(1.2,0.3)
    noise = np.random.normal(130,5)
    risk = 0
    data.append([rpm,temp,vib,noise,risk])
for _ in range(800):
    rpm = np.random.normal(27000,500)
    temp = np.random.normal(950,30)
    vib = np.random.normal(3.5,0.3)
    noise = np.random.normal(140,5)
    risk = 1
    data.append([rpm,temp,vib,noise,risk])
for _ in range(400):
    rpm = np.random.normal(20000,500)
    temp = np.random.normal(990,30)
    vib = np.random.normal(1.2,0.3)
    noise = np.random.normal(130,5)
    risk = 1
    data.append([rpm,temp,vib,noise,risk])
for _ in range(400):
    rpm = np.random.normal(20000,500)
    temp = np.random.normal(600,30)
    vib = np.random.normal(4.0,0.3)
    noise = np.random.normal(135,5)
    risk = 1
    data.append([rpm,temp,vib,noise,risk])
for _ in range(800):
    rpm = np.random.normal(28000,500)
    temp = np.random.normal(800,30)
    vib = np.random.normal(2.5,0.3)
    noise = np.random.normal(135,5)
    risk = 0
    data.append([rpm,temp,vib,noise,risk])
aircraft_data = pd.DataFrame(data,columns = ['Engine Rotation Speed','Engine Temperature','Engine Vibration','Noise','Risk'] )
aircraft_data = aircraft_data.sample(frac=1).reset_index(drop=True)
aircraft_data = aircraft_data.round(2)
aircraft_data.to_csv("aircraft_sensor_data.csv",index=False)
print(aircraft_data)
```
